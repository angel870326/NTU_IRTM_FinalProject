```python
from ckiptagger import WS, POS, NER
import csv
```

### 設定 CKIP

```python
ckip_dir = "data/"
ws, pos, ner = [f(ckip_dir) for f in [WS, POS, NER]]
```

### 檔案清理

```python
# covid_clean.csv：id(標題空白), content, date
# filePath = "../tokens+cvaw/covid_clean.csv"
# stock_clean.csv：id(標題空白), content, date
filePath = "../tokens+cvaw/stock_clean.csv"

fh = open(filePath, "r", encoding = "utf-8")  #讀檔
headerList = fh.readline().split(",")
csvList = fh.readlines()
fh.close()
```

```python
texts = []

for i in range(len(csvList)):
	csvList[i] = csvList[i].split(",")
	texts += [csvList[i][1]]
```

```python
print(texts[0])
```

```python
# 斷詞測試

word_s = ws(texts[0:2], sentence_segmentation=True,
	segment_delimiter_set = set('?？!！。.,，；:：、'))

print(word_s)
```

### CKIP 斷詞

```python
# word_s (list) 內存入每一篇(list in list)的斷詞結果


word_s = ws(texts, sentence_segmentation=True,
	segment_delimiter_set = set('?？!！。.,，；:：、'))

print("done")
```

```python
print(csvList[0][1])
print(csvList[1][1])
```

```python
# csvList 中的 content 用空白段開詞
for i in range(len(word_s)):
    csvList[i][1] = " ".join(word_s[i])

print(csvList[0][1])
print(csvList[1][1])
```

### 存檔

```python
# fw = open("covid_CKIPtoken.csv", "w", encoding = "utf-8")  #存檔
fw = open("stock_CKIPtoken.csv", "w", encoding = "utf-8")  #存檔
```

```python
fw.write(",".join(headerList))

for i in range(len(word_s)):
    fw.write(",".join(csvList[i]))
```

```python
fw.close()
```
